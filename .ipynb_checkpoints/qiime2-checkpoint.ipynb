```python
%%file Snakefile

rule all:
    input:
        'ASVs/denoising_stats.qzv',
        'ASVs/taxonomy.qzv',
        'ASVs/rarefaction_curve.qzv',
        'core-metrics/unweighted_unifrac_pcoa_results.qza',
        'core-metrics/weighted_unifrac_pcoa_results.qza',
        'core-metrics/faith_pd_vector.qza',
        'core-metrics/observed_otus_vector.qza',
        'core-metrics/shannon_vector.qza'
rule import_merged:
    input:
        directory('merged')
    output:
        'demux-merged.qza'
    shell:
        [
            "qiime tools import "
            "  --type 'SampleData[SequencesWithQuality]' "
            "  --input-path {input} "
            "  --input-format CasavaOneEightSingleLanePerSampleDirFmt "
            "  --output-path {output}"
        ][0]
        
rule dada2_denoised_single:
    input:
        'demux-merged.qza'
    output:
        'ASVs/denoising_stats.qza',
        'ASVs/table.qza',
        'ASVs/representative_sequences.qza'
    threads: 14
    shell:
          [  
            "qiime dada2 denoise-single "
            "  --i-demultiplexed-seqs {input} "
            "  --p-trunc-len 0 "
            "  --p-max-ee 5.0 "
            "  --p-n-threads {threads} "
            "  --output-dir ASVs"
          ][0]
            
rule tabulate_denoised:
    input:
        'ASVs/denoising_stats.qza'
    output:
        'ASVs/denoising_stats.qzv'
    shell:
        [
            "qiime metadata tabulate "
            "    --m-input-file {input} "
            "    --o-visualization {output}"
        ][0]

        
rule get_classifier:
    output:
        'silva-132-99-515-806-nb-classifier.qza'
    shell:
        'wget https://data.qiime2.org/2019.10/common/silva-132-99-515-806-nb-classifier.qza'
        
rule classify:
    input:
        'silva-132-99-515-806-nb-classifier.qza',
        'ASVs/representative_sequences.qza'
    output:
        'ASVs/taxonomy.qza'
    shell:
        [
            'qiime feature-classifier classify-sklearn '
                '--i-classifier {input[0]} '
                '--i-reads {input[1]} '
                '--o-classification {output}'
        ][0]
        
rule filter_features:
    input:
        'ASVs/table.qza',
        'ASVs/taxonomy.qza'
    output:
        'ASVs/filtered_table.qza',
        'ASVs/filtered_table.qzv',
    shell:
        [
            'qiime taxa filter-table '
            '  --i-table {input[0]} '
            '  --i-taxonomy {input[1]} '
            '  --p-exclude Mitochondria,Chloroplast,Eukaryota '
            '  --o-filtered-table temp.filter_features.qza && '

            'qiime feature-table filter-features '
            '  --i-table temp.filter_features.qza '
            '  --p-min-frequency 500 '
            '  --p-min-samples 2 '
            '  --o-filtered-table {output[0]} && '
            
            'qiime feature-table summarize '
            '  --i-table {output[0]} '
            '  --o-visualization {output[1]} &&'
            
            'rm temp.filter_features*'
        ][0]
           
rule tabulate_classifications:
    input:
        'ASVs/taxonomy.qza'
    output:
        'ASVs/taxonomy.qzv'
    shell:
        [
            'qiime metadata tabulate '
                '--m-input-file {input} '
                '--o-visualization {output}'
        ][0]
        

rule tree:
    input:
        'ASVs/representative_sequences.qza'
    output:
        aligned='ASVs/aligned_representative_sequences.qza',
        masked='ASVs/masked-aligned-rep-seqs.qza',
        unrooted='ASVs/unrooted-tree.qza',
        rooted='ASVs/rooted-tree.qza'
    shell:
        [
            'qiime phylogeny align-to-tree-mafft-fasttree '
                '--i-sequences {input} '
                '--o-alignment {output.aligned} '
                '--o-masked-alignment {output.masked} '
                '--o-tree {output.unrooted} '
                '--o-rooted-tree {output.rooted}'
        ][0]
        
rule alpha_rarefaction:
    input:
        table='ASVs/filtered_table.qza',
        tree='ASVs/rooted-tree.qza',
    output:
        'ASVs/rarefaction_curve.qzv'
    shell:
        [
            'qiime diversity alpha-rarefaction '
                '--i-table {input.table} '
                '--p-max-depth 9000 '
                '--i-phylogeny {input.tree} '
                '--m-metadata-file SampleMetadata.tsv '
                '--p-steps 20 '
                '--o-visualization {output}'
        ][0]
        
rule core_metrics:
    input:
        tree='ASVs/rooted-tree.qza',
        table='ASVs/filtered_table.qza',
        meta='SampleMetadata.tsv'
    output:
        'core-metrics/unweighted_unifrac_pcoa_results.qza',
        'core-metrics/weighted_unifrac_pcoa_results.qza',
        'core-metrics/faith_pd_vector.qza',
        'core-metrics/observed_otus_vector.qza',
        'core-metrics/shannon_vector.qza'
    shell:
        [
            'qiime diversity core-metrics-phylogenetic '
                '--i-phylogeny {input.tree} '
                '--i-table {input.table} '
                '--p-sampling-depth 6900 '
                '--m-metadata-file {input.meta} '
                '--output-dir core-metrics  '
        ][0]
```

```python
from sklearn import manifold
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style("white")

matrix_files = [
    'beta-diversity/weighted-unifrac-north_matrix.qza',
    'beta-diversity/unweighted-unifrac-north_matrix.qza',
    'beta-diversity/weighted-unifrac-south_matrix.qza',
    'beta-diversity/unweighted-unifrac-south_matrix.qza'
]

fig, axes = plt.subplots(nrows=2,ncols=2,figsize=(8,8),tight_layout=False,dpi=300)

i = [0,0,0,1,1,0,1,1]

for qza in matrix_files:

    a = !unzip $qza
    digest = a[1].split('/')[0].replace('  inflating: ','')
    inf = digest + '/data/distance-matrix.tsv'
    matrix = pd.read_table(inf,index_col=0)
    !rm -r $digest
    
    basin = qza.split('-')[-1].split('_')[0]
    distance = qza.split('/')[-1].rpartition('-')[0]
    metadata = pd.read_table('fish_samples_metadata_%s_swabs.tsv' % basin)
    metadata = metadata.loc[metadata['#SampleID'].isin(matrix.columns)]
    mds = manifold.MDS(max_iter=1000,n_init=1000,n_jobs=10,dissimilarity='precomputed',random_state=4783)
    mds.fit(matrix)
    pos = mds.embedding_
    pos = pd.DataFrame(data=pos,columns=['x','y'],index=matrix.index)
    ax = axes[i.pop(0)][i.pop(0)]
    for sp in set(metadata.SP):
        data = pos.loc[pos.index.isin(metadata.loc[metadata['SP'] == sp]['#SampleID'])]
        sns.scatterplot(x='x',y='y',data=data, ax=ax, label=sp,legend=False)
        ax.set_title(basin + ' ' + distance)
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    ax.set_xlabel('')
    ax.set_ylabel('')
```
